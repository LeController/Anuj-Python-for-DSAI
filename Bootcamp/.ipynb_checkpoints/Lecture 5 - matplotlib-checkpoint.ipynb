```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.available
```

```python
#Task 1
#1 Given$$y = 2x - 5 + 20 * noise$$, where noise is value between 0 and 1, generate x with 500 values between 0 and 10. Then generate the corresponding y.
import random
import numpy as np
import matplotlib.pyplot as plt


noise = np.random.random(500)
print("Noises =", noise[:10])
x = 10*np.random.random(500)
print( "x =", x[:10])
y = 2*x - 5 + 20*noise
```

```python
#2 Plot the above with a scatter plot.
plt.scatter(x,y)
```

```python
#3 Plot a line $y = 2x - 5$ on top of this scatter plot.
y_real = 2*x - 5
mpl.pyplot.plot(x,y_real)
plt.scatter(x,y)
```

```python
#3 Given a function called sigmoid with $$y = 1 / (1 + np.exp(-x))$$, generate x as points from -5 to 5, then generate the corresponding y. Plot the scatter graph. Set the title to "Sigmoid" function, and set the xlabel to "x", and ylabel to "y"
x = np.linspace(-5,5,29)
y = 1/(1+ np.exp(-x))
plt.scatter(x,y)
plt.title("Sigmoid")
plt.xlabel('x')
plt.ylabel('y')
```

```python
#5 Go to sklearn.datasets and research about make_classification API. Using this API, generate 500 samples, with n_features=10. Then plot the scatter graph of the first two features with color equals to their class, marker = 'o', size = 25, and edgecolor = black.
from sklearn import datasets
X, y = datasets.make_classification(n_samples=500, n_features=10)
plt.scatter(X[:, 0], X[:,1])
```

```python
#6 Load the iris datasets from sklearn. Set X as the first two features, and y as the target. Plot them in a scatter plot with color as their class. Add some grid to the plot.
iris = datasets.load_iris()
X = iris.data[:, :2]

y = iris.target

plt.scatter(X[:,0],X[:,1], label="class 0", c=y)
plt.title("IRIS")
plt.grid()
```
