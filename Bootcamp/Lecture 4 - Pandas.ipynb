```python
#6 Since Hepatatis B has a lot of nans, and highly correlate with Diptheria, drop column Hepatatis. Also drop column Population since there are way too many nans
df.drop(columns = ['Hepatitis B', "Population"], inplace=True)
```

```python
#7 Convert Status to 0 or 1. (0 for Developing and 1 for Developed)
labels, _ = pd.factorize(df["Status"])
df["Status"] = labels
```

```python
#8 Rename column thinness 1-19 years to thinness 10-19 years
df.rename(columns={"thinness 1-19 years":"thinness 10-19 years"}, inplace=True)
```

```python
#9 Get all columns except Life Expectancy, and convert to numpy array called X
X_all = df.drop(columns=['Country',"Life expectancy "]).values.astype(float)
X_all.shape
```

```python
#10 Get the column life expetancy and convert to numpy array called y
y = df['Life expectancy '].values.astype(float)
y
```

```python
#Task 2
#1 Continuing "howlongwelive.csv", check whether which column has how many missing data

print(np.sum(df.isnull()))
```

```python
#2 Fix all missing data using means
df.fillna(df.mean(), inplace=True)
```

```python
#3 Perform a groupby country. Which country has the lowest/highest average life expectancy?
groupCountry = df.groupby('Country')
series = groupCountry.mean()['Life expectancy ']
print(series.sort_values())

print("Thailand = ",series.loc['Thailand'])
```

```python
#4 Perform a groupby status. Is there any strong difference between developed and developing countries in their life expectancy?
groupStatus = df.groupby('Status')
series = groupStatus.mean()['Life expectancy ']
series
```

```python
#5 Create another dataframe manually with 2 columns. First column is the ID column with same value as country column. Also add another column Noise_level, and populate with random values (whatever you like).
countries = np.unique(df['Country'])
noises = np.random.uniform(50,100,(countries.shape))
noise_df = pd.DataFrame({'Country': countries, 'Noise Level': noises})
noise_df
```

```python
#6 Merge the two datasets together based on the ID column.
df = df.merge(noise_df, on='Country')
df
```

```python
#Task 3
#1 Load "data/appl_1980_2014.csv" into df
df2 = pd.read_csv("Data/appl_1980_2014.csv")
df2.columns
```

```python
#2 Transform Date column to datetime type

df2["Date"] = pd.to_datetime(df2["Date"])
```

```python
#3 Then filter only with year not less than 1987
df2[df2['Date'].dt.year > 1986]
```

```python
#4 Sort the index in an ascending order (oldest date first) - use sort_index
df2 = df2.set_index('Date')
df2.sort_index(ascending=True)
```

```python
#5 We want to know how apple does on end of every month, thus get the mean value for each columns, resample at last buiness day of each month (i.e., BM)
df2_months = df2.resample('BM').mean()
```

```python
#6 BTW, how many months do we have in our data?
len(df2_months["Volume"])
```

```python
#7 Compare the differences between this year and last year High column, using shift(), where the differences is simply this_year - last_year. Plot this graph, where x-axis is date, and y-axis is the gain/loss
df2_high = df2['High']
df2_high_last_year = df2_high.shift(365)
df2_high.plot(style=['g-'], label='This Year')
df2_high_last_year.plot(style=['b-'], label='Last Year')
diff = df2_high - df2_high_last_year
diff.plot(style=['r-'], label='Difference')
import matplotlib.pyplot as plt
plt.legend()
```

```python
#8 Perform a rolling mean (moving average) of Close with window size of 100 days
df2_close = df2['Close']
rolling = df2_close.rolling(100).mean().plot(style='-', label='rolling')
df2_close.plot(style='--',figsize=(10,6), label='normal')
plt.title("ffff")
plt.ylabel('Close Price')
plt.legend()
```
