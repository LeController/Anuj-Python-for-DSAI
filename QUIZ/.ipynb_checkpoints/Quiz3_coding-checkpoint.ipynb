```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report

X, y = make_blobs(n_samples=200, centers=3,
                  random_state=0, cluster_std=1.0)
```

```python
figure = plt.figure()
ax = plt.axes()
ax.grid()
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# 1. Prepare your data
#standardize
scaler = StandardScaler()
X = scaler.fit_transform(X)

#do train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
class KNN:
    def __init__(self, distance, k=3):
        self.k = k
        self.distance = distance
    
    # 2. Function for pairwise distance
    def find_distance(self, X_train, X_test):
        #create newaxis simply so that broadcast to all values
        dist = X_test[:, np.newaxis, :] - X_train[np.newaxis, :, :]
        sq_dist = dist ** 2

        #sum across feature dimension, thus axis = 2
        summed_dist = sq_dist.sum(axis=2)
        sq_dist = np.sqrt(summed_dist)
#         print('sq_dist.shape:',sq_dist.shape)
        return sq_dist
    
    # 3. Argosrt the pairwise distance matrix either with or without k
    def find_neighbors(self, X_train, X_test):
        if self.distance == 'euclidean':
            dist = self.find_distance(X_train, X_test)
            neighbors_ix = np.argsort(dist)[:, 0:self.k]
            return neighbors_ix
        elif self.distance == 'manhattan':

            dist = abs(X_test[:, np.newaxis, :] - X_train[np.newaxis, :, :])

            #sum across feature dimension, thus axis = 2
            summed_dist = sq_dist.sum(axis=2)
            dist = np.sqrt(summed_dist)
#         print('sq_dist.shape:',sq_dist.shape)

#             print("DIST SHAPE:", dist.shape)
#             dist = manhattan_distance(X_train, X_test)
            neighbors_ix = np.argsort(dist)[:, 0:self.k]
            return neighbors_ix
        else:
            raise ValueError("Distance must be euclidean or manhattan!")
        return neighbors_ix
    
    def find_neighbors_without_k(self, X_train, X_test):
        dist = self.find_distance(X_train, X_test)
        #return ALL neighbors
        neighbors_ix = np.argsort(dist)
#         print('WITHOUT k neighbors_ix.shape:',neighbors_ix.shape)
        return neighbors_ix

    # 4. Get the majority class 
    # If the majority class of the first place is equal to the second place, then ask the algorithm to pick the next nearest neighbors as the decider
    def get_most_common(self, y, n_classes, X_train, X_test):
        # k nearest neighbors
        y_nearest = y[0:self.k]
#         print('y_nearest:',y_nearest,'with k=',self.k)
        
        # returns a list of the counts of each digit up to max number of classes
        bincount = np.bincount(y_nearest, minlength=n_classes)
#         print('bincount:', bincount)
        largest = bincount.argmax()
        second_largest = bincount.argsort()[::-1][1]
#         print('largest:', largest)
#         print('second_largest:', second_largest)
    
        if bincount[largest] == bincount[second_largest]:
            # add one neighbor
            y_nearest = y[0: self.k+1]
            return np.bincount(y_nearest).argmax(), bincount[largest] / bincount.sum()
        return np.bincount(y_nearest).argmax(), bincount[largest] / bincount.sum()
    
    # implement cross validation
    def cv(self, X_train, y_train, cv, k):
        foldsize = int(X_train.shape[0]/cv)
        yhat_cv = np.zeros((len(k), cv))
        yhat_cv_prob = np.zeros((len(k), cv))
        
        for k_idx, kneighbors in enumerate(k):
            self.k = kneighbors
            for fold_idx, i in enumerate(range(0, X_train.shape[0], foldsize)):
                X_test_ = X_train[i:i+foldsize]
                y_test_ = y_train[i:i+foldsize]
                X_train_ = np.concatenate((X_train[:i], X_train[i+foldsize:]))
                y_train_ = np.concatenate((y_train[:i], y_train[i+foldsize:]))
                yhat, yhat_prob = self.predict(X_train_, X_test_, y_train_)
                accuracy = np.sum(yhat == y_test_)/len(y_test_)
                yhat_cv[k_idx, fold_idx] = accuracy
                yhat_cv_prob[k_idx, fold_idx] = yhat_prob.mean()
        return yhat_cv, yhat_cv_prob
    
    def predict(self, X_train, X_test, y_train):
        n_classes = len(np.unique(y_train))
#         print('Number of classes:', n_classes)
        # choose whether you want with k or not
        neighbors_ix = self.find_neighbors(X_train, X_test)
        yhat = np.zeros(X_test.shape[0])
#         print('initial yhat:', yhat.shape)
        yhat_prob = np.zeros(yhat.shape[0])
#         print('yhat_prob:',yhat_prob.shape)
#         print('y_train[neighbors_ix]:',y_train[neighbors_ix][0])
        for ix, y in enumerate(y_train[neighbors_ix]):
#             print('y:',y)
            yhat[ix], yhat_prob[ix] = self.get_most_common(y, n_classes, X_train, X_test)
#         print('yhat:\n',yhat, yhat.shape)
#         print('yhat_prob:\n', yhat_prob, yhat_prob.shape)
        return yhat, yhat_prob
```

```python
#ASSUMING we're doing this with K (as opposed to without K)

model = KNN(k=3, distance='euclidean')
yhat = yhat_prob = model.predict(X_train, X_test, y_train)
```

```python
# check best k from 1 to 15
k = np.arange(1,16)
cv_score, prob_score = model.cv(X_train, y_train, 10, k)
```

```python
for i, k_value in enumerate(k):
    print(f"Score with k={k_value}: ", cv_score.mean(axis=1)[i], "and prob_score of ", prob_score.mean(axis=1)[i])
```
