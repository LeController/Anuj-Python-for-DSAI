```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.stats import multivariate_normal
import math
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1500, cluster_std=[1.0, 3.5, 0.5], 
                  random_state=42)

#define basic params
m, n = X.shape
K = 3
max_iter = 20
oldNLL = 0
```

```python
#==initialization==

#responsibliity
r = np.full(shape=(m, K), fill_value=1/K)
print('r:\n',r, r.shape)

#pi
pi = np.full((K, ), fill_value=1/K) #simply use 1/k for pi
print('\npi:\n',pi, pi.shape)

#mean
random_row = np.random.randint(low=0, high=m, size=K)
mean = np.array([X[idx,:] for idx in random_row ]).T #.T to make to shape (M, K)
print('\nmean:\n', mean, mean.shape)

#covariance
cov = np.array([np.cov(X.T) for _ in range (K)])
print('\ncov:\n', cov, cov.shape)
```

```python
for iteration in range(max_iter):
    
    #===E-Step=====
    #Update r_ik of each sample
    for i in range(m):
        for k in range(K):
            xi_pdf = multivariate_normal.pdf(X[i], mean=mean[:, k], cov=cov[k])
            r[i, k] = pi[k] * xi_pdf  #p(y) * p(x | y)
        r[i] /= np.sum(r[i])  #normalize
        
#     print("Shape of r: ", r.shape)  #likelihood for each sample for each cluster
#     print("r: ", r)
    NK = np.sum(r, axis=0)
    assert NK.shape == (K, )  #sum of all likelihood of each cluster
#     print(NK)
#     print("Pi (before): ", pi)
    pi = NK / m
    assert pi.shape == (K, )
#     print("Pi (after): ", pi)
    mean =  ( X.T @ r ) / NK
    assert mean.shape == (n, K)
    
    for k in range(K):
        for i in range(m):
            X_mean = (X[i]-mean[:, k]).reshape(-1, 1)
            cov[k] += r[i, k] * (X_mean @ X_mean.T)
        cov[k] /= NK[k]
    assert cov.shape == (K, n, n)
    
    
    NLL = 0 #negative log likelihood
    for i in range(m):
        for k in range(K):
            NLL += math.log(pi[k])
            NLL += multivariate_normal.logpdf(
                X[i], mean=mean[:, k], cov=cov[k])
    
    #every 5 iteration plot once
    if(iteration % 5 == 0):
        preds = np.argmax(r, axis=1)
        plt.figure()
        plt.scatter(X[:, 0], X[:, 1], c=preds)
        plt.title(f"Iteration: {iteration}")
        print("NLL smaller? ", NLL)
        print("NLL difference: ", math.fabs(NLL - oldNLL)) # float absolute
        
    if(math.fabs(NLL-oldNLL) < 5):
        print("Final NLL: ", NLL)
        print("final difference: ", math.fabs(NLL - oldNLL))
        break
    oldNLL = NLL
        
```
